# Lats/lons?

I went down a wormhole reading about latitudes and longitudes re MISR data. Ultimately there are two ways of getting that info, (1) from pre-made files that relate the block number, line and sample numbers to the latitude and longitude, or (2) by calculating it based on the block number, line and sample numbers and the block offsets. I couldn't see an obvious way to get the block offsets from the netCDF file... but after some Googling, I found that I could download an Ancillary Geographic Product file from the main MISR download page (e.g. where I got the aerosol data) - there are some different ones, but I want the MIANCAGP product. I've downloaded the corresponding file to the aerosol product above (order number: 629083485).

# AOD and Angstrom for MODIS OC processing?

So I've got the data visualized, the real question remains: what value should I use for AOD and the Angstrom coefficient in the processing of the MODIS ocean color data?

## AOD

The AOD in MISR data is given at 550nm, the one from ocean colour data is at ~869nm. However, the MISR Angstrom is "calculated using the AODs at 550 and 86 nm", therefore I can use MISR AOD and Angstrom to determine AOD-860:

```python
aod860 = ((550/860)**ang[:])*aod[:]
```

```python
lonE = 165
lonW = 161
latN = 57
latS = 54

plt.figure(figsize=(9,5))
ax = plt.axes(projection=ccrs.PlateCarree())

#doing the gridlines, haven't figured out how to
#do the xticks neatly, so for now, I'm keeping the 
#grey gridlines on the image
gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=1, color='gray', alpha=0.5, linestyle='--')
gl.xlabels_top = False
gl.ylabels_right = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER

#adding land
land = cfeature.GSHHSFeature(
        scale='l',lw=2,
        facecolor=[0.9,0.9,0.9])
ax.add_feature(land)

# Set lat and long limits here
ax.set_extent([lonE,lonW,latS,latN])
ax.set_aspect('equal')

#plotting the satellite data
plt.pcolor(lon, lat, np.ma.masked_where(np.isnan(aod860),aod860), 
             transform=ccrs.PlateCarree(),zorder=2,
              vmin=0,vmax=0.34)
plt.colorbar(label='AOD860')
plt.savefig('MISR_AOD860_zoom.png',dpi=350)
```

### subscening

Let's look at the data in the plume region, which is defined by the following limits (me visually selecting):

```python
lonE = 164
lonW = 162.5
latN = 56
latS = 55.3

lonData = lon[:].data
latData = lat[:].data

lonMask = np.ma.masked_where((lonData>lonE)| (lonData<lonW),lonData).mask
latMask = np.ma.masked_where((latData>latN)| (latData<latS),latData).mask
latlonMask = np.ma.mask_or(lonMask,latMask)
```

```python
aodData = aod[:]
aodData.data[latlonMask] = np.nan

angData = ang[:]
angData.data[latlonMask] = np.nan
```

### Wavelengths for AOD?

In the MODIS processing, I need to specify `taua` at all wavelengths (not including IR - according to Sean on the Forum):

```python
wls = np.asarray([412,443,469,488,531,547,555,645,667,678,748,859,869,1240,1640,2130])
aod_array = ((wls/550.)**np.expand_dims(angData,axis=2))*np.expand_dims(aodData,axis=2)
aod_mean = np.nanmean(aod_array, axis=(0,1))
```
